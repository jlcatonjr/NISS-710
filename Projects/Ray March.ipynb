```python
import matplotlib.pyplot as plt

def formatted_scatter_matrix(data, pp = False):  
    # Create a figure showing scatterplots given in scatter_cats  
    fig_len = 15  
    fig, ax = plt.subplots(figsize = ((fig_len, fig_len)))  
    # Use fig_len to dictate fig_size, adjust size of font, size of dots, etc...  
    num_vars = len(data.keys())  
    fontsize = 65 / num_vars  
    plt.rcParams.update({'font.size': fontsize})  
    pd.plotting.scatter_matrix(data,alpha = .5, s = 108 / num_vars, ax=ax)  
    # tight layout improves layout of text and plots in the figure  
    plt.tight_layout()  
    plt.show()  
    if pp != False:
        pp.savefig(fig, bbox_inches = "tight")  
    plt.close()

scatter_keys = list(data.keys())[3:-2]
formatted_scatter_matrix(data[scatter_keys])
```

```python
fips = sorted(set(list(data["fips"])))
state_codes = {
    'WA': '53', 'DE': '10', 'DC': '11', 'WI': '55', 'WV': '54', 'HI': '15',
    'FL': '12', 'WY': '56', 'PR': '72', 'NJ': '34', 'NM': '35', 'TX': '48',
    'LA': '22', 'NC': '37', 'ND': '38', 'NE': '31', 'TN': '47', 'NY': '36',
    'PA': '42', 'AK': '02', 'NV': '32', 'NH': '33', 'VA': '51', 'CO': '08',
    'CA': '06', 'AL': '01', 'AR': '05', 'VT': '50', 'IL': '17', 'GA': '13',
    'IN': '18', 'IA': '19', 'MA': '25', 'AZ': '04', 'ID': '16', 'CT': '09',
    'ME': '23', 'MD': '24', 'OK': '40', 'OH': '39', 'UT': '49', 'MO': '29',
    'MN': '27', 'MI': '26', 'RI': '44', 'KS': '20', 'MT': '30', 'MS': '28',
    'SC': '45', 'KY': '21', 'OR': '41', 'SD': '46'
}

for state, code in state_codes.items():
    code = int(code)
    data["fips"][data["fips"] == code] = state
plot_data = data.set_index(["fips"])
plot_data
```

```python
plt.rcParams.update({"font.size": 25}) 
dim = int(len(fips) ** .5) + 1
ob_min = data["obesity_"].min()
ob_max = data["obesity_"].max()
fig, ax = plt.subplots(dim - 1, dim, figsize = ((dim - 1) * 5, dim * 5))
states = list(state_codes.keys())
states.sort()
for i in range(dim - 1):
    for j in range(dim):
        
        try:
            state_num = i * (dim) + j
            state = states[state_num]
            plot_data.loc[state].plot.scatter(x = "Year", y = "obesity_", ax = ax[i][j], s = 100)
            ax[i][j].set_ylim(ob_min, ob_max)
            ax[i][j].set_title(state)
            if j > 0:
                ax[i][j].set_ylabel("")
                ax[i][j].set_yticks([])
            if i < (dim - 1):
                ax[i][j].set_xlabel("")
                ax[i][j].set_xticks([])
                
        except:
                ax[i][j].set_ylabel("")
                ax[i][j].set_yticks([])
                ax[i][j].set_xlabel("")
                ax[i][j].set_xticks([])
            
        continue
```
