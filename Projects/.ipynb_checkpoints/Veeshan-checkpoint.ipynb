```python
import pyreadstat
data, meta = pyreadstat.read_dta("2017_Bahunepati_forreg.dta")
```

```python
stats_vars = ["householdsize","hhdage", "agriland_valuer",
              "dist_market",
              "partn_microf", "housedamage", "proptdamage",
             "married", "hindu", "headEduc", "BrahminChhetri",
             "occu_agri", "Female", "injuries"]
```

```python
import pandas as pd
import statsmodels
from scipy import stats
import numpy as np
import pingouin as pg
from scipy.stats import ttest_ind



distances = ["dist0", "dist10"]#, "dist30", "dist60", "dist90", "dist120"]
results = {}
for key in data[stats_vars]:
    results[key] = {}
    for i in range(1, len(distances)):
        name1 = distances[i-1]
        name2 = distances[i]

        inside1 = data[data[name1] == 1][key]
        outside1 = data[data[name1] == 0][key]
        inside2 = data[data[name2] == 1][key]
        outside2 = data[data[name2] == 0][key]

        inside1_n = inside1.count()
        outside1_n = outside1.count()
        inside2_n = inside2.count()
        outside2_n = outside2.count()
        
        results[key]["inside1_mean"] = inside1.mean()
        results[key]["outside1_mean"] = outside1.mean()
        results[key]["inside2_mean"] = inside2.mean()
        results[key]["outside2_mean"] = outside2.mean()
        results[key]["inside1_sem"] = inside1.sem()
        results[key]["outside1_sem"] = outside1.sem()
        results[key]["inside2_sem"] = inside2.sem()
        results[key]["outside2_sem"] = outside2.sem()
        
        results[key]["inside1_t"] = results[key]["inside1_mean"] / results[key]["inside1_sem"]
        results[key]["outside1_t"] = results[key]["outside1_mean"] / results[key]["outside1_sem"]
        results[key]["inside2_t"] = results[key]["inside2_mean"] / results[key]["inside2_sem"]
        results[key]["outside2_t"] = results[key]["outside2_mean"] / results[key]["outside2_sem"]
        results[key]["t_val1"], results[key]["p_val1"] = \
            ttest_ind(inside1.dropna(), outside1.dropna())                
        results[key]["t_val2"], results[key]["p_val2"] = \
            ttest_ind(inside2.dropna(), outside2.dropna())                

pd.set_option('display.max_columns', None)

pd.DataFrame(results).T
        
    
```
