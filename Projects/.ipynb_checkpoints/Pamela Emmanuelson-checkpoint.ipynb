# Import data, save turn A and turn B as separateentries in a dictionary

```python
import pandas as pd
data_dct = {}
data_dct["A"] = pd.read_excel("network.xlsx", sheet_name="A", index_col = [0])
data_dct["B"] = pd.read_excel("network.xlsx", sheet_name="B", index_col = [0])
```

# Save results in a dictionary by turn and session.

```python
results={}
for session in range(1,9):
    results[session]={}
    for turn in ["A", "B"]:
        results[session][turn] = {}
        for period in range(1,4):
            name = "Session "+ str(session) + " Per" + str(period)
            results[session][turn][period] = data_dct[turn][name]            
```

# Test whether or not turn order causes a statistically significant change of mean.

```python
for session in range(1,9):
    for turn in ["A", "B"]:
        results[session][turn] = pd.DataFrame(results[session][turn])
                
results
```

```python
import statsmodels
from scipy import stats
import numpy as np
import pingouin as pg
from scipy.stats import ttest_ind

for session in range(1,9):
    # check that difference of means for turn order isnot  statistically significant
    session_test = ttest_ind(results[session]["A"].stack(), results[session]["B"].stack())
    print("session: " + str(session),
          "t-stat: " + str(round(session_test[0],3)),
          "p-val: " + str(round(session_test[1], 3)), 
          "",
          sep = "\n" )
```
